# qgis-bridge Demo

Push spatial data from Python directly into a running QGIS session.

**Prerequisites:**
- QGIS Desktop is open
- The qgis-bridge plugin is installed and enabled in QGIS

## 1. Basic Vector Layer

Create a simple GeoDataFrame and push it to QGIS with default styling.

```python
import geopandas as gpd
from shapely.geometry import Point
from qgis_bridge import to_qgis

# Create some sample points
gdf = gpd.GeoDataFrame(
    {"name": ["Office", "Park", "Station"], "value": [10, 25, 40]},
    geometry=[Point(-73.99, 40.74), Point(-73.97, 40.77), Point(-73.98, 40.75)],
    crs="EPSG:4326",
)

to_qgis(gdf, layer_name="Sample Points")
```

## 2. Color by a Numeric Column (Graduated)

Use `color_by` with a numeric column to get a graduated color ramp.

```python
to_qgis(gdf, layer_name="Points by Value", color_by="value", color_ramp="Viridis")
```

## 3. Color by a Categorical Column

Use `color_by` with a string column to get distinct colors per category.

```python
to_qgis(gdf, layer_name="Points by Name", color_by="name", color_ramp="Blues")
```

## 4. Polygons with Opacity

Works with any geometry type — points, lines, polygons.

```python
from shapely.geometry import box

zones = gpd.GeoDataFrame(
    {"risk_score": [1.2, 3.5, 7.8, 9.1]},
    geometry=[
        box(-73.99, 40.74, -73.98, 40.75),
        box(-73.98, 40.74, -73.97, 40.75),
        box(-73.99, 40.75, -73.98, 40.76),
        box(-73.98, 40.75, -73.97, 40.76),
    ],
    crs="EPSG:4326",
)

to_qgis(zones, layer_name="Risk Zones", color_by="risk_score", color_ramp="Reds", opacity=0.6)
```

## 5. Using the Accessor

The `.qgis` accessor is a shorthand — same thing, different syntax.

```python
zones.qgis.send(layer_name="Risk Zones (accessor)", color_by="risk_score")
```
